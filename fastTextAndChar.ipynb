```python
import codecs
data = []
classes=[]
with codecs.open('data.json') as f:
    big_json = json.load(f)
    for hyp in big_json:
        ents = hyp["entities"]
        text = hyp["text"]
        for ent in ents:
            text = text.replace(ent["value"], ent["type"])
        data.append(text)
        classes.append(hyp["intent"])
```

```python
word_index = tok.word_index
embed_size = 300
embedding_matrix = np.zeros((n_words, embed_size))
for word, i in word_index.items():
    if i >= n_words: continue
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
    else:
        embedding_matrix[i] = np.random.normal(embed_size)
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.layers import Dense, Input, LSTM, Embedding, Dropout, Activation
from keras.layers import Bidirectional, GlobalMaxPool1D
from keras.models import Model
from keras.initializers import Constant
from keras import initializers, regularizers, constraints, optimizers, layers
```

```python
word_in = Input(shape=(max_len,))
x = Embedding(n_words, embed_size, embeddings_initializer=Constant(embedding_matrix), input_length=max_len, trainable=True)(word_in)
x = Bidirectional(LSTM(50, return_sequences=True, dropout=0.1, recurrent_dropout=0.1))(x)
x = GlobalMaxPool1D()(x)
x = Dense(50, activation="relu")(x)
x = Dropout(0.1)(x)
x = Dense(22, activation="softmax")(x)
model = Model(inputs=word_in, outputs=x)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_word, y_shuf, validation_split=0.1, batch_size=32, epochs=30, verbose=1)
```

```python
X_char = []
for sentence in sentences_shuf:
    sent_seq = []
    for i in range(max_len):
        word_seq = []
        for j in range(max_len_char):
            try:
                word_seq.append(char2idx.get(sentence[i][j]))
            except:
                word_seq.append(char2idx.get("PAD"))
        sent_seq.append(word_seq)
    X_char.append(np.array(sent_seq))
```

```python
word_in = Input(shape=(max_len,))
emb_word = Embedding(n_words, embed_size, embeddings_initializer=Constant(embedding_matrix), input_length=max_len, trainable=True)(word_in)

# input and embeddings for characters
char_in = Input(shape=(max_len, max_len_char,))
emb_char = TimeDistributed(Embedding(input_dim=n_chars + 2, output_dim=10,
                           input_length=max_len_char, mask_zero=True))(char_in)
# character LSTM to get word encodings by characters
char_enc = TimeDistributed(LSTM(units=20, return_sequences=False,
                                recurrent_dropout=0.3))(emb_char)
# main LSTM
x = concatenate([emb_word, char_enc])
x = SpatialDropout1D(0.3)(x)
main_lstm = Bidirectional(LSTM(units=50, return_sequences=False,
                               recurrent_dropout=0.3))(x)
out = Dense(22, activation="softmax")(main_lstm)

model = Model([word_in, char_in], out)

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["acc"])

history = model.fit([X_word,
                     np.array(X_char).reshape((len(X_char), max_len, max_len_char))],
                    y_shuf,
                    batch_size=32, epochs=250, validation_split=0.1, verbose=1)
```
